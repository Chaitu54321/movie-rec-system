```python
import streamlit as st
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import re
```

```python
df=pd.read_csv("C:\\Users\\CHAITANYA\\Documents\\MOVIE REC SYS\\movies.csv")
df
```

```python
df = df[df["genres"].notnull()]
df = df[df["genres"].str.lower() != "(no genres listed)"]
df = df[df["genres"].str.lower() != "no genres listed"]

df["year"] = df["title"].apply(lambda x: int(re.findall(r'\((\d{4})\)', x)[0]) if re.search(r'\((\d{4})\)', x) else None)
df["clean_genres"] = df["genres"].str.replace("|", " ")
df
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean_genres"])
model = NearestNeighbors(metric="cosine", algorithm="brute")
model.fit(X)
```

**UI**

```python
import joblib
```

```python
joblib.dump(vectorizer, "vectorizer.pkl")
joblib.dump(model, "movie_model.pkl")
df.to_csv("clean_movies.csv", index=False)

print("✅ Model and vectorizer saved successfully!")
```
